#### Import Packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

___
#### Read _Tips_ Dataset
___

```python
tips = sns.load_dataset('tips')
sns.set()
```

```python
tips.head()
```

___
Plot _tips_ Dataset
___ 

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)

```

### The bill amount decreases on Fridays.

```python
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], kind="point", data=tips)
```

### Smokers give a little bit more tips 

```python
tips["weekend"] = tips["day"].isin(["Fri", "Sat"])
sns.catplot(x="day", y="total_bill", hue="weekend", kind="box", dodge=False, data=tips)
```

```python
clean_plot = sns.catplot(x="day", y="total_bill", kind="violin", inner=None, data=tips)
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips, ax=clean_plot.ax)
```
