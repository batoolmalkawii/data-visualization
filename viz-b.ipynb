___
#### Import Packages
___

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

___
#### Read _diamonds_  Dataset
___

```python
diamonds = sns.load_dataset('diamonds')
sns.set()
```

```python
diamonds.head()
```

___
Plot _diamonds_ Dataset
___ 

```python
sns.catplot(x="color", y="price", kind="boxen", data=diamonds.sort_values("color"))
```

### The diamond price varies based on the color.

```python
sns.catplot(x="color", y="price", hue="cut", kind="violin", data=diamonds)
```

### Ideal-cut diamonds have the highest prices 

```python

sns.catplot(x="cut", y="depth", hue="color", kind="point", data=diamonds)
```

### Fair-cut Diamonds have the highest depths.
